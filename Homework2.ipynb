Bài 3 : Tính tam giác: Nhâp vào basố a,b,c bất kì. Kiểm tra xem nó có thể là
độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh
cua mot tam giac'. Ngượclại, kiểm tra xem nó là tam giác gì: Đều, Cân,
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giácvà in ra màn
hình.


```python
a = float(input())
b = float(input())
c = float(input())
if a + b > c and a + c > b and b + c > a:
    perimeter = a + b + c
    p = perimeter / 2
    s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    if a == b == c:
        print("Đều")
    elif a == b or b == c or a == c:
        if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
            print("Vuông Cân")
        else:
            print("Cân")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Vuông")
    else:
        print("Tam giác thường")
    print(s, perimeter)
else:
    print("Khong la ba canh cua mot tam giac")
```

```python
h = int(input())
for i in range(1, h + 1):
    spaces = " " * (h - i)
    if i == 1:
        print(spaces + "*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        inner_spaces = " " * (2 * i - 3)
        print(spaces + "*" + inner_spaces + "*")
```

không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25….


```python
a = int(input())
is_square = False
i = 1
while i * i <= a:
    if i * i == a:
        is_square = True
        break
    i += 1
print(is_square)
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không


```python
n = int(input())
is_power = True
if n <= 0:
    is_power = False
else:
    temp = n
    while temp > 1:
        if temp % 3 != 0:
            is_power = False
            break
        temp //= 3
print(is_power)
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A


```python
a = int(input())
f1 = 1
f2 = 1
while f2 <= a:
    temp = f1 + f2
    f1 = f2
    f2 = temp
print(f1)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
n = int(input())
count = 0
temp = n
while temp > 0:
    digit = temp % 10
    if digit == 2 or digit == 3 or digit == 5 or digit == 7:
        count += 1
    temp //= 10
print(count)
```
