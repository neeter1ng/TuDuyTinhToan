### W5A3
Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def prime(k):
    if k < 2:
        return False
    for i in range(2, int(k**0.5 + 1)):
        if k // i:
            return False
        return True
```

```python
def is_perfect_number(n):
    sum = 0
    for i in range(1, int(n**0.5 +1)):
        if n % i == 0:
            sum += i
    if sum == n:
        return True
    return False
```

### W5A8 
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

```python
def hamming_distance(n1, n2):
    xor = n1 ^ n2
    distance = 0
    while xor > 0:
        xor &= (xor - 1)
        distance += 1
    return distance
```

### W5A10
Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài.

```python
def dang_cau(a, b):
    dict_a = {}
    dict_b = {}
    for i in range(lens(a)):
        char_a = a[i]
        char_b = b[i]
        if char_a in dict_a:
            if dict_a[char_a] != dict_b:
                return False
        if char_b in dict_b:
            if dict_b[char_b] != dict_a:
                return False
        dict_a[char_a] == dict_b
        dict_b[char_b] == dict_a
    return True
```

### W5A12
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|.

```python
def divide_with_remainder(dividend, divisor):
    q = dividend // divisor
    r = dividend % divisor
    if dividend == divisor*q + r and 0 <= abs(r) < abs(divisor):
        return q, r
```

Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.


```python
def quadratic_roots_count(a, b, c):
    if a == 0:
        if b != 0:
            return 1
        if b == 0 and c != 0:
            return 0
        if b == 0 and c == 0:
            return -1
    delta = b**2 - 4*a*c
    if delta == 0:
        return 1
    if delta > 0:
        return 2
    return 0
```

### W5A25
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    (x1, y1), (x2, y2), (x3, y3), (x4, y4) = quad
    px, py = p
    cp1 = (x2 - x1) * (py - y1) - (y2 - y1) * (px - x1)
    cp2 = (x3 - x2) * (py - y2) - (y3 - y2) * (px - x2)
    cp3 = (x4 - x3) * (py - y3) - (y4 - y3) * (px - x3)
    cp4 = (x1 - x4) * (py - y4) - (y1 - y4) * (px - x4)
    if (cp1 >= 0 and cp2 >= 0 and cp3 >= 0 and cp4 >= 0) or (cp1 <= 0 and cp2 <= 0 and cp3 <= 0 and cp4 <= 0):
        return True
    return False
```

### W5A28 
Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math
def tinh_goc(x1,y1,x2,y2,x3,y3):
    ba_x, ba_y = x1 - x2, y1 - y2
    bc_x, bc_y = x3 -x2, y3- y2
    tich_vo_huong = ba_x * bc_x + ba_y * bc_y
    tich_do_dai = (ba_x**2 + ba_y**2)** 0.5 * (bc_x**2 + bc_y**2)**0.5
    cos = tich_vo_huong / tich_do_dai
    if cos < -1 or cos > 1:
        return False
    goc_rad = math.acos(cos)
    goc_deg = math.degrees(goc_rad)
    return goc_deg
```

### W5A33
Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
import math

def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        fare_0_8 = 11000
        if km <= 0.8: dist_fare = fare_0_8
        elif km <= 20: dist_fare = fare_0_8 + (km - 0.8) * 13000
        else: dist_fare = fare_0_8 + (20 - 0.8) * 13000 + (km - 20) * 10500
    else:
        fare_0_8 = 12000
        if km <= 0.8: dist_fare = fare_0_8
        elif km <= 20: dist_fare = fare_0_8 + (km - 0.8) * 14000
        else: dist_fare = fare_0_8 + (20 - 0.8) * 14000 + (km - 20) * 12000
    wait_fare = 0
    if wait_minutes > 5:
        blocks = math.ceil((wait_minutes - 5) / 5)
        wait_fare = blocks * 2000
    subtotal = dist_fare + wait_fare
    is_night = start_hour in [22, 23, 0, 1, 2, 3, 4, 5]
    if is_night:
        subtotal *= 1.2
    total = subtotal
    if km > 30 and not is_night:
        total *= 0.9
    if total < 50000 and km > 5:
        total = 50000
    return int(total)
```
